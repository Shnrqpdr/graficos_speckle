```python
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
```

```python
frames = []
for i in range(2517):
    frames.append(0)
```

```python
db = list(df['8. ★D (D with highest r²) for Dʙ'])
sigma = list(df['7. STANDARD DEVIATION (σ) for D for Dʙ'])
lac = list(df['83. LACUNARITY for Image (Ｌ) = ∑Λ₍ɢ₎/Ǥɴ ∀ ɢ∈Ǥ'])
```

```python
print(db)
print(sigma)
print(lac)
```

```python
# create figure and axis objects with subplots()
fig,ax = plt.subplots()
# make a plot
ax.plot(frames,
        db,
        color="red", 
        marker=".")
# set x-axis label
ax.set_xlabel("Frames", fontsize = 14)
# set y-axis label
ax.set_ylabel("DB",
              color="red",
              fontsize=14)
# twin object for two different y-axis on the sample plot
ax2=ax.twinx()
# make a plot with different y-axis using second axis object
ax2.plot(frames, sigma,color="blue",marker="H")
ax2.set_ylabel("Desvio Padrão",color="blue",fontsize=14)
plt.show()
# save the plot as a file
fig.savefig('tentando.jpg',
            format='jpeg',
            dpi=100,
            bbox_inches='tight')
```

```python
fig, ax1 = plt.subplots()

ax2 = ax1.twinx()
ax1.plot(frames, db, 'g-')
ax2.plot(frames, sigma, 'b-')

ax1.set_xlabel('Frames')
ax1.set_ylabel('DB', color='g')
ax2.set_ylabel('Desvio Padrão', color='b')

plt.show()
```
